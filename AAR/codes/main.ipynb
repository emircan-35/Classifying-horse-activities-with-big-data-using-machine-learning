```python
#get clean csv files
path='C:\\Users\\emirc\\Desktop\\AAR\\codes\\dataHandled\\' #modify the path if needed
dfAll=getCSV(path,'All')
dfBacardi=getCSV(path,'Bacardi')
dfDrieku=getCSV(path,'Drieku')
dfGaloway=getCSV(path,'Galoway')
dfHappy=getCSV(path,'Happy')
dfPatron=getCSV(path,'Patron')
dfZafir=getCSV(path,'Zafir')
```

```python
def getAllStatisticsByHorse():
    allStatisticsByHorse={}
    allStatisticsByHorse['Bacardi']=featureExtraction(dfBacardi)
    allStatisticsByHorse['Drieku']=featureExtraction(dfDrieku)

    allStatisticsByHorse['Galoway']=featureExtraction(dfGaloway)
    allStatisticsByHorse['Happy']=featureExtraction(dfHappy)

    allStatisticsByHorse['Patron']=featureExtraction(dfPatron)
    allStatisticsByHorse['Zafir']=featureExtraction(dfZafir)

    return allStatisticsByHorse
```

```python
dfStatisticsAll=featureExtraction(dfAll)
```

```python
statisticsOfAllHorses=getAllStatisticsByHorse()
```

```python
def storeLocallyStatistics():
    dfStatisticsAll.to_csv('C:\\Users\\emirc\\Desktop\\AAR\\codes\\statistics\\all.csv',index=False)
    for key in statisticsOfAllHorses:
        statisticsOfAllHorses[key].to_csv('C:\\Users\\emirc\\Desktop\\AAR\\codes\\statistics\\'+key+'.csv',index=False)
```
