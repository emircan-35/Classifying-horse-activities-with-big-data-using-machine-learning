```python
def plot(data):
    for label in LABELS:
        df=splitByActivity(data,label)
        aX=df['Ax'].to_numpy()
        aY=df['Ay'].to_numpy()
        aZ=df['Az'].to_numpy()
        aX=aX[0:101]
        aY=aY[0:101]
        aZ=aZ[0:101]
        plt.title(label+" activity over time")
        plt.ylabel("Accelaration")
        plt.xlabel("TIME")
        plt.plot(aX,color='red')
        plt.plot(aY,color='green')
        plt.plot(aZ,color='blue')
        plt.savefig('C:\\Users\\emirc\\Desktop\\AAR\\codes\\plots\\'+label+'.png')
```

```python
def splitByActivity(df, activity):
    return df[df['label'].isin([activity])]
```
